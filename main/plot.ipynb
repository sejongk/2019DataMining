```python
import matplotlib.pyplot as plt
import pandas as pd 
import mglearn
from sklearn.preprocessing import StandardScaler
```

```python
X_train = pd.read_csv('big_raw_X.csv', delimiter = ',')
Y_train = pd.read_csv('big_raw_Y.csv', delimiter = ',')
```

```python
X_train = X_train.drop('Unnamed: 0',axis = 1)
Y_train = Y_train.drop('Unnamed: 0',axis = 1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_train)
```

```python
print(kmeans.labels_)
```

```python
X_train
```

```python
mglearn.discrete_scatter(X_train.iloc[:, 0], X_train.iloc[:, 65], kmeans.labels_, markers='o')
```

```python
kmeans2 = KMeans(n_clusters=11)
kmeans2.fit(X_train)
mglearn.discrete_scatter(X_train.iloc[:, 0], X_train.iloc[:, 65], kmeans2.labels_, markers='o')
```

```python
mglearn.discrete_scatter(X_train.iloc[:, 3], X_train.iloc[:, 68], kmeans.labels_, markers='o')
```

```python
plt.plot(X_train.iloc[:, 0], X_train.iloc[:, 65])
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN()
clusters = dbscan.fit_predict(X)
```
